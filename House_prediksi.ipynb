## Outlier

```
data_o = ['median_house_value','housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
        'distance_to_center']

def check_outlier(data_eda, data_o):
    hasil_outlier = {}
    for col in data_o:
        q1 = data_eda[col].quantile(0.25)
        q3 = data_eda[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        outliers = data_eda[(data_eda[col] < lower_bound) | (data_eda[col] > upper_bound)]
        hasil_outlier[col] = {
            'jumlah_outliers' : outliers.shape[0],
            'Persentase Outlier': (outliers.shape[0] / data_eda.shape[0]) * 100
        }
    return pd.DataFrame(hasil_outlier).T
check_outlier(data_eda, data_o).sort_values(by='Persentase Outlier', ascending=False)
```

## modeling mencari best models

```
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Load data
data_ml = pd.read_csv('data_ml.csv')
data_ml

# Pisahkan fitur dan target
X = data_ml.drop(['median_house_value'], axis=1)
y = data_ml['median_house_value']

# Split data 80 20
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.2, random_state=2020)

# split lagi yang 80 menjadi 60 20
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.25, random_state=2020)


imputasi = X.select_dtypes(include='number').columns.tolist()
encode = ['ocean_proximity']

imputer_scaling = Pipeline([
    ('knn_imputer', KNNImputer(n_neighbors=5)),
    ('rb_scaling', RobustScaler())
])
encoding = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])

preprocessor = ColumnTransformer(transformers=[
    ('knn_imputer', imputer_scaling, imputasi),
    ('onehot', encoding, encode)
], remainder='passthrough')

# Model yang diuji
models = {
    'LinearRegression': LinearRegression(),
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=42),
    'RandomForestRegressor': RandomForestRegressor(random_state=42),
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=42),
    'LGBMRegressor': LGBMRegressor(random_state=42),
    'XGBRegressor': XGBRegressor(random_state=42)
}

# Simpan hasil
model_results = []

for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regression', model)
    ])

    pipeline.fit(X_train, y_train)
    y_val_pred = pipeline.predict(X_val)
    y_train_pred = pipeline.predict(X_train)

    # Validasi
    val_mse = mean_squared_error(y_val, y_val_pred)
    val_rmse = np.sqrt(val_mse)
    val_mae = mean_absolute_error(y_val, y_val_pred)
    val_r2 = r2_score(y_val, y_val_pred)

    # Training
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_rmse = np.sqrt(train_mse)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)

    model_results.append({
        'Model': name,
        'Train RMSE': train_rmse,
        'Train MAE': train_mae,
        'Train R2': train_r2,
        'Test RMSE': val_rmse,
        'Test MAE': val_mae,
        'Test R2': val_r2,
    })

# Tampilkan hasil
results_df = pd.DataFrame(model_results)
results_df = results_df.sort_values(by='Test R2', ascending=False)  # atau 'Test R2' untuk performa terbaik
results_df
```

## Evalauasi cross validation top 3 best models

```
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import KNNImputer
from sklearn.model_selection import KFold, cross_val_score

# Load data
data_ml = pd.read_csv('data_ml.csv')
data_ml

# Pisahkan fitur dan target
X = data_ml.drop(['median_house_value'], axis=1)
y = data_ml['median_house_value']

# Split data jadi Train (80%) dan Test (20%) â†’ Stratified
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=20
)

imputasi = X.select_dtypes(include='number').columns.tolist()
encode = ['ocean_proximity']

imputer_scaling = Pipeline([
    ('knn_imputer', KNNImputer(n_neighbors=5)),
    ('rb_scaling', RobustScaler())
])
encoding = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])

preprocessor = ColumnTransformer(transformers=[
    ('knn_imputer', imputer_scaling, imputasi),
    ('onehot', encoding, encode)
], remainder='passthrough')

## kandidat model
models = {
    'RandomForestRegressor': RandomForestRegressor(random_state=42),
    'LGBMRegressor': LGBMRegressor(random_state=42),
    'XGBRegressor': XGBRegressor(random_state=42)

}

# Ganti StratifiedKFold â†’ KFold (karena regresi)
cv = KFold(n_splits=5, shuffle=True, random_state=42)

model_results = []

for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regression', model)
    ])

    # Gunakan skor R2 (atau kamu bisa ganti dengan neg_mean_squared_error, dsb.)
    scores = cross_val_score(pipeline, X_train, y_train, scoring='r2', cv=cv, n_jobs=-1)

    model_results.append({
        'Model': name,
        'Fold-1': scores[0],
        'Fold-2': scores[1],
        'Fold-3': scores[2],
        'Fold-4': scores[3],
        'Fold-5': scores[4],
        'Mean R2': scores.mean(),
        'Std R2': scores.std()
    })

# Tampilkan hasil
result_df = pd.DataFrame(model_results).T
result_df
```

```
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor


# Ambil preprocessor dari pipeline
preprocessor = pipeline.named_steps['preprocessor']

# Transform X_train â†’ hasil = array
X_transformed = preprocessor.transform(X_train)

# Ambil nama-nama fitur hasil transformasi
# Numerik
num_features = preprocessor.named_transformers_['knn_imputer'].named_steps['rb_scaling'].get_feature_names_out(imputasi)

# Kategorikal
cat_features = preprocessor.named_transformers_['onehot'].named_steps['onehot'].get_feature_names_out(encode)

# Gabungkan
feature_names = np.concatenate([num_features, cat_features])

# Buat DataFrame untuk VIF
X_vif = pd.DataFrame(X_transformed, columns=feature_names)

# Hitung VIF
vif_df = pd.DataFrame()
vif_df["Feature"] = X_vif.columns
vif_df["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

# Tampilkan hasil
vif_df.sort_values(by="VIF", ascending=False).reset_index(drop=True)
```
